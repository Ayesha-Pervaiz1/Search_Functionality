```python
import os
import requests
import openpyxl
from openpyxl.drawing.image import Image as ExcelImage
from google.cloud import vision
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
from PIL import Image
from io import BytesIO

# Ensure the image storage directory exists
IMAGE_DIR = "Engagement_rings1"
os.makedirs(IMAGE_DIR, exist_ok=True)

def create_excel(file_path):
    """Creates an Excel file with gem carat metadata in Column A."""
    wb = openpyxl.Workbook()
    sheet = wb.active
    sheet.title = "EngagementRing"
    
    # Adding headers
    sheet.cell(row=1, column=1, value="Gem Carat")
    for col in range(2, 12):
        sheet.cell(row=1, column=col, value=f"Image {col-1}")  # Updated for images
    sheet.cell(row=1, column=12, value="Vision API Labels")  # Column L for labels
    
    # Gem Carat metadata
    gem_carat_values = [
        "Under 1 Carat", "0.5 Carat", "1 Carat", "2 Carat", "2.5 Carat", "3 Carat",
        "4 Carat", "5 Carat", "6 Carat", "7 Carat", "8 Carat", "9 Carat", "10 Carat",
        "- 0.25 Carat Each Side Stone", "- Approx 0.25 Carat Smalls", "- Approx 0.5 Carat Smalls",
        "- Approx 0.75 Carat Smalls", "- Approx 1 Carat Smalls", "- Approx 2 Carat Smalls"
    ]
    
    # Inserting data into the sheet
    for idx, value in enumerate(gem_carat_values, start=2):
        sheet.cell(row=idx, column=1, value=value)
    
    # Save the Excel file
    wb.save(file_path)
    print(f"Excel file created successfully: {file_path}")

def search_image(query, api_key, cse_id):
    """Search for images using Google Custom Search JSON API."""
    try:
        service = build("customsearch", "v1", developerKey=api_key)
        res = service.cse().list(q=query + " engagement ring", cx=cse_id, searchType="image", num=10).execute()
        if "items" in res:
            return [item["link"] for item in res["items"]]
    except HttpError as e:
        print(f"Google Search API Error: {e}")
    return []

def analyze_image(image_url):
    """Analyze image using Google Cloud Vision API."""
    try:
        client = vision.ImageAnnotatorClient()
        image = vision.Image()
        image.source.image_uri = image_url
        response = client.label_detection(image=image)
        labels = [label.description.lower() for label in response.label_annotations]
        return labels
    except Exception as e:
        print(f"Vision API Error: {e}")
        return []

def download_image(image_url, save_path):
    """Download and save an image if it's valid."""
    try:
        response = requests.get(image_url, timeout=10)
        if response.status_code == 200:
            img = Image.open(BytesIO(response.content))
            img.verify()  # Ensure it's a valid image
            img = Image.open(BytesIO(response.content))  # Reopen after verification
            img = img.convert("RGB")  # Convert to standard format
            img.save(save_path, format="JPEG")  # Save as JPEG
            return save_path
        else:
            print(f"Failed to download image: {image_url} (Status Code: {response.status_code})")
    except Exception as e:
        print(f"Error downloading or verifying image: {e} - URL: {image_url}")
    return None

def update_excel_with_images(file_path, api_key, cse_id):
    """Fetch images, analyze with Vision API, filter for engagement rings, and embed images in Excel."""
    wb = openpyxl.load_workbook(file_path)
    sheet = wb.active

    for row in range(2, sheet.max_row + 1):
        metadata = sheet.cell(row, 1).value
        if metadata:
            existing_images = [sheet.cell(row, col).value for col in range(2, 12)]  # Columns B to K
            if not any(existing_images):  # Avoid re-fetching if images are already present
                print(f"Searching images for: {metadata}")
                image_urls = search_image(metadata, api_key, cse_id)
                if image_urls:
                    labels = []
                    for col, image_url in enumerate(image_urls, start=2):
                        # Generate unique file name
                        image_filename = f"{metadata.replace(' ', '_')}_{col-1}.jpg"
                        image_path = os.path.join(IMAGE_DIR, image_filename)

                        # Download and validate image
                        saved_image = download_image(image_url, image_path)

                        if saved_image:
                            print(f"Inserting image: {saved_image}")
                            try:
                                # Insert image into Excel
                                img = ExcelImage(saved_image)
                                img.width, img.height = 100, 100  # Resize images to fit cells
                                col_letter = chr(64 + col)  # Convert col number to letter (B, C, D, etc.)
                                sheet.add_image(img, f"{col_letter}{row}")
                            except Exception as e:
                                print(f"Error inserting image: {e} - File: {saved_image}")

                            # Analyze image using Vision API
                            img_labels = analyze_image(image_url)
                            labels.append(", ".join(img_labels))

                    # Store Vision API labels in column L
                    sheet.cell(row, 12, value=" | ".join(labels))
                    print(f"Images and labels for '{metadata}' added successfully!")

    wb.save(file_path)
    print("Excel updated with engagement ring images and labels!")

# Google API credentials
API_KEY = "AIzaSyAo5M5HXAXSkVxBzHRG0bUMRJf3RltEhJw"  # Replace with your actual API key
CSE_ID = "d63ff1141028241de"  # Replace with your actual CSE ID

# Set Google Cloud credentials
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = r"D:\Techionik_2025\Search_functionality\jewelry-449211-b1561756ae8d.json"  

# Excel file
EXCEL_FILE = "gem_carat_data_ERImage1.xlsx"

# Run the functions
create_excel(EXCEL_FILE)
update_excel_with_images(EXCEL_FILE, API_KEY, CSE_ID)
```
