```python
import requests
import base64
import json

# Replace with your actual API key
API_KEY = "AIzaSyBFMXfAz9O6sy8IOMTJhkl4qK4R6Xuux2Y"  

# Path to the image you want to analyze
image_path = r"D:\Techionik_2025\Search_functionality\Images\image_13.jpg"  # Use raw string for the path

# Read and encode the image in base64
with open(image_path, "rb") as image_file:
    encoded_image = base64.b64encode(image_file.read()).decode("utf-8")

# Google Vision API endpoint
url = f"https://vision.googleapis.com/v1/images:annotate?key={API_KEY}"

# Request payload
payload = {
    "requests": [
        {
            "image": {"content": encoded_image},
            "features": [{"type": "LABEL_DETECTION", "maxResults": 5}],  # Detects objects in the image
        }
    ]
}

# Send POST request
response = requests.post(url, json=payload)

# Parse and display the response
if response.status_code == 200:
    result = response.json()
    print("Labels detected:")
    for label in result["responses"][0]["labelAnnotations"]:
        print(f"- {label['description']} (Confidence: {label['score']:.2f})")
else:
    print(f"Error: {response.status_code}, {response.text}")
```
